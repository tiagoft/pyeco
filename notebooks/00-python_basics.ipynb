## Nested loops

```python
for i in range(4):
    for j in range(4):
        print(f"i={i}, j={j}")
```

```python
# Problem: find all prime numbers between 0 and N_MAX
N_MAX = 20
for i in range(2,N_MAX):
    i_is_prime = True
    for j in range(2,i):
        if i % j == 0:
            print(f"{i} is divisible by {j}")
            i_is_prime = False
            break
    if i_is_prime:
        print(f"{i} is prime")
```

```python
# Another solution, using a growing list of primes and append
N_MAX = 20
primes = [2, 3]
for i in range(4,N_MAX):
    i_is_prime = True
    for j in primes:
        if i % j == 0:
            i_is_prime = False
            break
    if i_is_prime:
        primes.append(i)

for i in primes:
    print(f'{i} is prime')
```

## Defining a function

```python
def next_integer(i):
    return i + 1

print(next_integer(5))

```

```python
def is_prime(i):
    for j in range(2,i):
        if i % j == 0:
            return False
    return True

print(is_prime(41))
```

```python
# Calling functions from withing functions:
def get_all_primes_up_to(N_MAX):
    ans = []
    for i in range(2,N_MAX):
        if is_prime(i):
            ans.append(i)
    return ans
```

```python
# Type hints : they help the reader and the linter to understand your code!
def get_all_primes_up_to(N_MAX: int) -> list[int]:
    ans: list[int] = []
    for i in range(2,N_MAX):
        if is_prime(i):
            ans.append(i)
    return ans

```

```python
# Docstrings: documentation for functions
def is_prime(i: int) -> bool:
    """[hey, this is our documentation!]Returns True if i is a prime number, False otherwise.

    Args:
        i (int): The integer to check for primality.
    Returns:
        bool: True if i is prime, False otherwise.
    """
    for j in range(2,i):
        if i % j == 0:
            return False
    return True

help(is_prime)
```
